## using `[]`

When passing a string or a list of strings to `[]`, a subset of columns is selected from the DataFrame.

````python
# select a Series (single column) of a DataFrame
df["u_company_name"]
````

````python
# Note that the result is a Pandas-Series, not a DataFrame!
type(df["u_company_name"])
````

````python
# Multiple columns are selected by supplying a list:
df[["u_company_name", "u_year"]]
````

````python
# The result is a DataFrame!
type(df[["u_company_name", "u_year"]])
````

````python
df.columns
````

````python
# lists created by code are especially useful for column selection

# Example: selecting columns starting with "cb_"
selected_cols = []
for c in df.columns:
    if c.startswith("cb_"):
        selected_cols.append(c)
print(f"selected_cols: {selected_cols}")

df[selected_cols]
````

````python
# Alternative solutions for generating the same `selected_cols`

# List comprehension:
selected_cols = [c for c in df.columns if c.startswith("cb_")]
print(f"From list comprehension: {selected_cols}")

# Using methods included in pandas:
selected_cols = df.columns[df.columns.str.startswith("cb_")]
print(f"Using pandas methods: {selected_cols}")
````

# Slice DataFrame rows using `[]`

When passing an integer range, or a list/array of integer or booleans to `[]`, the respective rows are selected.

[https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#slicing-ranges](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#slicing-ranges)

````python
# slice rows (exclusive end)
df[0:1]
````

````python
# Please note that the following will not work because pandas will look for a column with 0 as label when receiving only one integer:
# df[0]
````

````python
# select the first 10 rows
df[:10]
````

````python
# select the last 10 rows
df[-10:]
````

Please not that the integers provided to `[]` refer to **row numbers, not indices**! (this is different from using `.loc[]` and more like using `.iloc[]`)

Passing **booleans** will also select rows (not columns) when used with `[]`

````python
# note that the number of booleans passed to `[]` must equal the number of rows!
# df[[True, False, True]] => ValueError: Item wrong length 3 instead of 824.
````

````python
# Therefore, booleans used with `[]` are typically generated by code, e.g.:
mask = df["u_year"] == 2005
df[mask].head()
````

````python
# You can also use more complex boolean expressions to select rows that fulfill certain conditions
df[(df["u_year"] == 2005) & (df["u_iso3"] == "DEU")].head()
````

# Select rows and columns using the `df.loc[]` method

`.loc[]` selects rows by their indices and and columns by their label. Alternatively, booleans are accepted.

## Select rows using ```loc```

### Using indices

````python
df.loc[[0, 15, 27]]
````

### Using Booleans

````python
df.loc[df["u_year"] == 2005].head()
````

## Select columns using ```loc```

````python
df.loc[:, ["u_year", "u_company_name"]]
````

## Select rows and columns using `loc`

````python
df.loc[df["u_year"] > 2007, ["u_company_name", "u_year"]]
````
