## Challenge 1: Clean character names

**Goal:** Start with a messy list of names and produce a clean list.

The solution uses four steps:

1. remove leading and trailing whitespace with `strip()`
2. standardize capitalization with `title()`
3. remove duplicates with a `set`
4. sort the result with `sorted()`

```python
messy_names = [
    " luke skywalker ",
    "Leia Organa",
    "luke skywalker",
    " YODA ",
    "darth vader ",
    "leia organa",
]

cleaned_names = {name.strip().title() for name in messy_names}
final_names = sorted(cleaned_names)

print(final_names)
print(len(final_names))
```

The set comprehension creates one cleaned version of each name. Since sets
keep unique values only, repeated names disappear automatically.

The final call to `sorted()` converts the set back into a sorted list.

## Challenge 2: Classify movie episodes

**Goal:** Loop through a dictionary of movie episodes and classify each
episode number.

The solution uses an `if` / `elif` / `else` block. Episodes 1 to 3 are the
prequel trilogy, episodes 4 to 6 are the original trilogy, and everything
else is classified as other.

```python
episodes = {
    1: "The Phantom Menace",
    2: "Attack of the Clones",
    3: "Revenge of the Sith",
    4: "A New Hope",
    5: "The Empire Strikes Back",
    6: "Return of the Jedi",
    7: "The Force Awakens",
}

for episode_number, title in episodes.items():
    if 1 <= episode_number <= 3:
        group = "Prequel trilogy"
    elif 4 <= episode_number <= 6:
        group = "Original trilogy"
    else:
        group = "Other"

    print(f"Episode {episode_number}: {title} ({group})")
```

We can also use conditions inside a list comprehension to collect only the
original trilogy titles.

```python
original_trilogy_titles = [
    title for episode_number, title in episodes.items() if 4 <= episode_number <= 6
]

print(original_trilogy_titles)
```

A dictionary comprehension can create a new dictionary that maps each episode
number to its group.

```python
episode_groups = {}
for episode_number in episodes:
    if 1 <= episode_number <= 3:
        episode_groups[episode_number] = "Prequel trilogy"
    elif 4 <= episode_number <= 6:
        episode_groups[episode_number] = "Original trilogy"
    else:
        episode_groups[episode_number] = "Other"

print(episode_groups)
```

## Challenge 3: Write reusable helper functions

**Goal:** Turn repeated logic into reusable functions.

The first function cleans a list of names. It uses the same logic as the
collections challenge, but packages it in a function so that it can be reused.

```python
def clean_names(names: list[str]) -> list[str]:
    """Clean, deduplicate, and sort character names.

    Args:
        names: Raw names that may contain extra whitespace, inconsistent
            capitalization, and duplicates.

    Returns:
        Cleaned names sorted alphabetically.

    """
    cleaned_names = {name.strip().title() for name in names}
    return sorted(cleaned_names)


print(clean_names(messy_names))
```

The next function classifies one episode number. Having this logic in a
function makes it easy to apply it to many episode numbers.

```python
def classify_episode(episode_number: int) -> str:
    """Return the trilogy group for a Star Wars episode number.

    Args:
        episode_number: Star Wars episode number.

    Returns:
        The trilogy group label for the episode.

    """
    if 1 <= episode_number <= 3:
        return "Prequel trilogy"
    if 4 <= episode_number <= 6:
        return "Original trilogy"
    return "Other"


for episode_number in [1, 4, 5, 7]:
    print(episode_number, classify_episode(episode_number))
```

We can now use the function in a dictionary comprehension.

```python
episode_groups = {
    episode_number: classify_episode(episode_number) for episode_number in episodes
}

print(episode_groups)
```

The final helper function converts fractions into formatted percentage
strings. The number of decimal places is controlled by an argument with a
default value.

```python
def make_percent_strings(values: list[float], decimals: int = 1) -> list[str]:
    """Convert fractions to formatted percentage strings.

    Args:
        values: Fractions to convert, such as 0.125 for 12.5%.
        decimals: Number of decimal places in the formatted percentages.

    Returns:
        Percentage strings with the requested number of decimal places.

    """
    return [f"{value * 100:.{decimals}f}%" for value in values]


fractions = [0.0, 0.125, 0.2, 0.755, 1.0]
print(make_percent_strings(fractions))
print(make_percent_strings(fractions, decimals=2))
```

## Combined example

This final cell shows all helper functions together. This is useful as a
quick check that the functions work on the example inputs.

```python
episode_numbers = [1, 4, 5, 7]

print(clean_names(messy_names))
print([classify_episode(number) for number in episode_numbers])
print(make_percent_strings(fractions, decimals=1))
```
